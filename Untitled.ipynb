```python
# Read images and steering angles from driving_dataset folder
from __future__ import division
import os
import numpy as np
from scipy import pi
from itertools import islice

# Define the data folder and training file
DATA_FOLDER = '/Users/adarshmac/Desktop/driving_dataset'  # Change this to your folder
TRAIN_FILE = os.path.join(DATA_FOLDER, 'data.txt')

# Define the split ratio for training and validation
split = 0.8

# Lists to store image paths and steering angles
X = []
y = []

# Define the limit of lines to read (if needed)
LIMIT = None  # Set to None to read all lines, or specify a number

# Read the data file
with open(TRAIN_FILE) as fp:
    for line in islice(fp, LIMIT):
        path, angle = line.strip().split()
        full_path = os.path.join(DATA_FOLDER, path)
        X.append(full_path)
        # Convert angle from degrees to radians
        y.append(float(angle) * pi / 180)

# Convert lists to numpy arrays for easier handling
y = np.array(y)
print("Completed processing data.txt")

# Split the data into training and validation sets
split_index = int(len(y) * split)
train_y = y[:split_index]
test_y = y[split_index:]

print(f"Total samples: {len(y)}")
print(f"Training samples: {len(train_y)}")
print(f"Validation samples: {len(test_y)}")
```

```python
import numpy;

# PDF of train and test 'y' values. 
import matplotlib.pyplot as plt 
plt.hist(train_y, bins=50, normed=1, color='green', histtype ='step');
plt.hist(test_y, bins=50, normed=1, color='red', histtype ='step');
plt.show()
```

```python
# Model 0: Baseline Model: y_test_pred = mean(y_train)
train_mean_y = np.mean(train_y)

# Calculate Mean Squared Error (MSE) for the baseline model
mse_mean = np.mean(np.square(test_y - train_mean_y))
mse_zero = np.mean(np.square(test_y - 0.0))

print('Test_MSE(MEAN): %f' % mse_mean)
print('Test_MSE(ZERO): %f' % mse_zero)
```
